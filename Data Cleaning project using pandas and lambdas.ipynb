```python
#here we are importing the library pandas and we read excel file 
import pandas as pd
df = pd.read_excel("/Users/revanthchowdary/Desktop/Projects/jupyterProjects/Data sets/Customer Call List.xlsx")
df
```

```python
# here we see that from above we have some values which are duplicate ones as we are planning for the data cleaning we just need to do 
# remove the duplicate values using the duplicate cmd
# df.drop_duplicates()
df = df.drop_duplicates()
df
```

```python
# Here we are going to do drop the coloumns which are not useful for us and then store the values to variable df
df = df.drop(columns="Not_Useful_Column")
df
```

```python
# Now as a dataanalyst we just need to do remove the unvanted values from the data as we found some in the last_name
# now here we use strip from left then its like LSTRIP
# we gave 123, ., _, / as they are unwanted

df["Last_Name"] = df["Last_Name"].str.lstrip("123./_")
df
```

```python
# we now after that we see that phone_numbers are very wrong so we make automation to this to clean the numbers
# we remove non-numeric values from the phonenumber coloumn we use the replace with the above

df['Phone_Number'] = df['Phone_Number'].str.replace('[^A-Za-z0-9]', '', regex=True)

df['Phone_Number'] = df['Phone_Number'].apply(lambda x : str(x))

df['Phone_Number'] = df['Phone_Number'].apply(lambda x : x[0:3] + '-' + x[3:6] + '-' + x[6:10])

df['Phone_Number'] = df['Phone_Number'].str.replace('--', '')

df['Phone_Number'] = df['Phone_Number'].str.replace('--', '')

df
```

```python
# now we just got each and everything with phone numbers above
# Nows just make the paying customer values as perfectly assigned

df['Paying Customer'] = df['Paying Customer'].str.replace('Yes', 'Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No', 'N')

# Now lets do work on the Do_Not_Contact row which was also not clear

df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes', 'Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No', 'N')
```
